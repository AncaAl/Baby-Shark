```python
#define function for visual view

def print_with_hist(pd_series, size=12):
    min = 1000000000
    max = 0
    max_result_str_len = 0
    for i, k in pd_series.items():
        result_str_len = len(f"{i} {' '*(2-(len(str(i))-1))} {k}")
        if result_str_len > max_result_str_len:
            max_result_str_len = result_str_len
        if k > max:
            max = k
        if k < min:
            min = k
    
    for i, k in pd_series.items():
        result_str = f"{i} {' '*(2-(len(str(i))-1))} {k}"
        print(
            "MAX ->" if k == max else "MIN ->" if k == min else "      ", 
            result_str, 
            " " * (max_result_str_len - len(result_str)), 
            " | ",
            "*" * int((k / size)),
        )
```
